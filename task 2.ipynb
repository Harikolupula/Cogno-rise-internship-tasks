```python
Manufacturer_of_cereal={'A': "American Home Food Products",
"G":"General Mills",
"K":"Kelloggs",
"N":"Nabisco",
"P":"Post",
"Q":"Quaker Oats",
"R":"Ralston Purina"}
df["mfr"]=df["mfr"].map(Manufacturer_of_cereal)
```

```python
plt.figure(figsize=[5,5])
sns.countplot(x='mfr',data=df)
plt.xlabel("mfr")
plt.ylabel("Count by mfr")
plt.xticks(rotation=45,ha='right')
```

```python
type1={'C':'cold',
'H':'Hot'}
df["type"]=df["type"].map(type1)
```

```python
plt.figure(figsize=[3,3])
sns.countplot(x='type',data=df)
plt.xlabel("type")
plt.ylabel("Count of type")
```

```python
plt.figure(figsize=[5,15])
plt.barh(df['name'],df['rating'])
plt.xlabel('Rating')
plt.ylabel('Name')
```

```python
pd.crosstab(df['sugars'],['rating']).plot(kind='line',stacked=True)
plt.xlabel('Sugars')
plt.ylabel('Rating')
```

```python
plt.scatter(df['calories'],df['rating'])
plt.title("Relationship")
plt.xlabel("calories")
plt.ylabel("rating")
```

```python
pd.crosstab(df['protein'],['rating']).plot(kind='line',stacked=True)
plt.xlabel('Protein')
plt.ylabel('Rating')
```

```python
plt.figure(figsize=[10,5])
sns.heatmap(df[['calories','protein','fat','sodium','fiber',
       'carbo','sugars','potass','vitamins','shelf','weight','cups',
       'rating']].corr(),annot=True)
```
